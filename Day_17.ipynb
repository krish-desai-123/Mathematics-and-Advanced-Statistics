4Ô∏è‚É£ Provide real-world interpretation.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

lambda_val = 8

prob_10 = stats.poisson.pmf(10, lambda_val)
print(f'Probability of 10 cases: {prob_10}')
```

```python
 x = np.arange(0,17)
 probs = stats.poisson.pmf(x, lambda_val)
 plt.plot(x, probs, marker = '.', color = 'white')
 plt.scatter(x[10], prob_10, marker = 'D', color = 'red', label = 'Prob of 10')
 plt.legend()
 plt.title('Poisson Distribution')
 plt.xlabel('Cases per hour')
 plt.ylabel('Probability')
 plt.show()
```

2Ô∏è‚É£ For normal distribution (Mean=100, Std=15):
- Compute probability score above 130.
- Find Z-score.
- Plot PDF and CDF.

```python
mean = 100
std = 15
x = 130

z_score = (x - mean) / std
prob_130 = stats.norm.pdf(z_score)
prob_130
```

```python
x_vals = np.linspace(50,160, 1000)
prob_pdf = stats.norm.pdf(x_vals, mean, std)

plt.plot(x_vals, prob_pdf)
plt.axvline(x=x, color = 'red')
plt.title('Normal Distribution')
plt.xlabel('Score')
plt.ylabel('Probability')
plt.show()
```

```python
cdf_vals = stats.norm.cdf(x_vals, mean, std)
plt.plot(x_vals, cdf_vals)
plt.axvline(x=x, color = 'red')
plt.title('Cumulative Distribution')
plt.xlabel('Score')
plt.ylabel('Probability')
plt.show()
```
