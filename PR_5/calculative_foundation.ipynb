```python
import numpy as np
from scipy.linalg import lu
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

#### Part A: Vector & Matrix Fundamentals

```python
data = np.array([
    [85, 90, 88, 92],
    [60, 55, 65, 70],
    [95, 85, 92, 88],
    [45, 50, 40, 55],
    [78, 80, 85, 82]
])
data
```

```python
s1 = data[0]
s2 = data[1]
```

```python
norm1_s1 = np.linalg.norm(s1, ord=1)
norm2_s1 = np.linalg.norm(s1, ord=2)
```

```python
dot_product = np.dot(s1, s2)
```

```python
cos_theta = dot_product / (np.linalg.norm(s1) * np.linalg.norm(s2))
angle_deg = np.degrees(np.arccos(cos_theta))
```

```python
s1_3d = s1[:3]
s2_3d = s2[:3]
cross_product = np.cross(s1_3d, s2_3d)
```

```python
projection_s1_s2 = (np.dot(s1, s2) / np.dot(s2, s2)) * s2
```

```python
print(f"Norm-1 of Student 1: {norm1_s1}")
print(f"Norm-2 of Student 1: {norm2_s1:.2f}")
print(f"Dot Product (Student 1 & 2): {dot_product}")
print(f"Angle between Student 1 & 2: {angle_deg:.2f} degrees")
print(f"Cross Product (First 3 subjects): {cross_product}")
print(f"Projection of Student 1 onto Student 2: {projection_s1_s2}")
```

#### Part B: Matrix Operations

```python
A = data
B = data + 5
```

```python
matrix_addition = A + B

matrix_multiplication = np.dot(A.T, A)

matrix_transpose = A.T

matrix_inverse = np.linalg.inv(matrix_multiplication)
matrix_determinant = np.linalg.det(matrix_multiplication)
```

```python
print("Matrix Transpose:\n", matrix_transpose)
print("Determinant of A.T * A:", matrix_determinant)
```

#### Part C: Linear Transformations & Geometry

###### A line (1D) connects two data points (students), while a plane (2D) is a flat surface that can connect multiple points.

###### A hyperplane is simply a geometric boundary used to divide data, and it always has exactly one dimension less than your total dataset space.

###### As dimensions increase with more subjects, the divider scales up: a 1D line divides a 2D space, a 2D plane divides a 3D space, and a 3D hyperplane divides a 4D space.

#### Part D: Eigenvalues & Decomposition

```python
cov_matrix = np.cov(A)
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

P, L, U = lu(A)

U_svd, S_svd, Vt_svd = np.linalg.svd(A)
```

```python
print("Eigenvalues:\n", eigenvalues)
print("\nLU Decomposition (L Matrix):\n", L)
print("\nSVD (Singular Values):\n", S_svd)
```

#### Part E: Dimensionality Reduction

```python
pca = PCA(n_components=2)
A_pca = pca.fit_transform(A)

student_averages = np.mean(A, axis=1)
overall_mean = np.mean(student_averages)
labels = np.where(student_averages >= overall_mean, "Above Average", "Below Average")

lda = LDA(n_components=1)

A_lda = lda.fit_transform(A, labels)
```

```python
print("PCA:\n", A_pca)
print("\nDummy Classification Labels:\n", labels)
print("\nLDA Transformation:\n", A_lda)
```
