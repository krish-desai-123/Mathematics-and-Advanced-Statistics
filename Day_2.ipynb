```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.Series([50000,73000,76000,85000,77000,45000,65000,55000,71000,50000,60000,73000])

print(f'Mean: {np.mean(df)}')
print(f'Median: {np.median(df)}')
print(f'Mode: {df.mode()[0]}')
```

```python
print(f'Range: {df.max()-df.min()}')
print(f'Variance: {np.var(df):.2f}')
print(f'Standard Deviation: {np.std(df):.2f}')
```

```python
df_with_outlier = df.tolist() + [1000000]
df_with_outlier = pd.Series(df_with_outlier)

print('DF with outlier stats: ')
print(f'Mean: {np.mean(df_with_outlier)}')
print(f'Median: {np.median(df_with_outlier)}')
print(f'Mode: {df_with_outlier.mode()[0]}')
```

```python
sns.histplot(data=df,kde=True, color='white')
plt.axvline(np.mean(df),color='red',linestyle='--',label='Mean')
plt.axvline(np.median(df),color='blue',linestyle='--',label='Median')
plt.title('Histogram of normal df')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

df.plot(kind='box',color='red')
plt.title('Boxplot of normal df')
plt.xlabel('Sales')
```

```python
sns.histplot(data=df_with_outlier,kde=True, color='white',bins=20)
plt.axvline(np.mean(df_with_outlier),color='red',linestyle='--',label='Mean')
plt.axvline(np.median(df_with_outlier),color='blue',label='Median',linestyle='--')
plt.title('Histogram of df with outlier')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()

sns.boxplot(data=df_with_outlier,color='red',orient='h')
plt.title('Boxplot of df with outliers')
plt.xlabel('Sales')
```
