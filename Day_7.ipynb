## ðŸ“Œ Mini Assignment

Spam filter case study:
- 20% emails are spam
- P(Detected | Spam) = 0.95
- P(Detected | Not Spam) = 0.10

Calculate P(Spam | Detected).

```python
import numpy as np
import pandas as pd

np.random.seed(9)
email = pd.DataFrame({
    'Detected' : np.random.choice(['Yes','No'], size=1000),
    'Type' : np.random.choice(['Spam','Not Spam'], size=1000)
})

email.head()
```

```python
prob_spam = len(email[email['Type']=='Spam']) / len(email)
print(f'Probability of spam: {prob_spam}')
```

```python
prob_not_spam = len(email[email['Type']=='Not Spam']) / len(email)
print(f'Probability of not spam: {prob_not_spam}')
```

```python
prob_type = email['Type'].value_counts(normalize=True)
prob_type
```

```python
prob_detected = len(email[email['Detected']=='Yes']) / len(email)
print(f'Probability of detected: {prob_detected}')
```

```python
prob_not_detected =  len(email[email['Detected']=='No']) / len(email)
print(f'Probability of not detected: {prob_not_detected}')
```

```python
prob_detected_table = email['Detected'].value_counts(normalize=True)
prob_detected_table
```

```python
prob_detected_spam = len(email[(email['Detected'] == 'Yes') & (email['Type'] == 'Spam')]) / len(email)
print(f'Probability of being detected and is spam: {prob_detected_spam}')
```

```python
prob_detected_not_spam = len(email[(email['Detected'] == 'Yes') & (email['Type'] == 'Not Spam')]) / len(email)
print(f'Probability of being detected and is not spam: {prob_detected_not_spam}')
```

```python
prob_not_detected_spam = len(email[(email['Detected'] == 'No') & (email['Type'] == 'Spam')]) / len(email)
print(f'Probability of not being detected and is spam: {prob_not_detected_spam}')
```

```python
prob_not_detected_not_spam = len(email[(email['Detected'] == 'No') & (email['Type'] == 'Not Spam')]) / len(email)
print(f'Probability of not being detected and is not spam: {prob_not_detected_not_spam}')
```

```python
contingency_table = pd.crosstab(email['Detected'], email['Type'])
contingency_table
```
