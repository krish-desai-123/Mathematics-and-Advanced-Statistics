## ðŸ“Œ Mini Assignment
Generate website visits data and analyze mean & variance.

```python
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.DataFrame({
    'Day' : range(1,366),
    'Traffic' : np.random.randint(70,300,365),
})
df.head()
```

```python
pmf = df['Traffic'].value_counts(normalize=True).sort_index()
pmf
```

```python
expected_value = sum(pmf.index * pmf.values)
print(f'Expected Value: {expected_value:.2f}')
```

```python
variance_value = sum(((pmf.index - expected_value) ** 2) * pmf.values)
print(f'Variance Value: {variance_value:.2f}')
```

```python
print(f'Std Value: {variance_value ** (1/2):.2f}')
```
