
## ðŸ“Œ Mini Assignment

Generate salary data of 1000 employees.

Tasks:
1. Check normality visually.
2. Compute skewness and kurtosis.
3. Interpret results.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

np.random.seed(42)
df = pd.Series(np.random.randint(30000,70000,1000))
df.head()
```

```python
df.describe()
```

```python
sns.histplot(df,kde=True,bins=30)
plt.axvline(df.mean(),color='r',linestyle='--',label='Mean')
plt.axvline(df.median(),color='b',linestyle='--',label='Median')
plt.axvline(df.mode()[0],color='y',linestyle='--',label='Mode')
plt.legend()
plt.show()
```

```python
skewness = skew(df)

if -0.5 <= skewness <= 0.5:
    print(f'Skewness is {skewness:.2f}: Symmetric. No transformation needed.')

elif -1 <= skewness <= 1:
    print(f'Skewness is {skewness:.2f}: Moderately skewed. Monitoring suggested.')
else:
    print(f'Skewness is {skewness:.2f}: Highly skewed. Transformation is needed.')
```

```python
kurto = kurtosis(df)

if -0.30 <= kurto <= 0.30:
    print(f'Kurtosis is {kurto:.2f}: Mesokurtic')
elif kurto < -0.30:
    print(f'Kurtosis is {kurto:.2f}: Platykurtic')
else:
    print(f'Kurtosis is {kurto:.2f}: Leptokurtic')
```

```python
'''
    This data doesn't fit the normal distribution but it is symmetric with skewness of 0.07. It is platykurtic meaning less outlier risk.
'''
```
