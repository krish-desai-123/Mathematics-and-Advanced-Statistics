```python
import numpy as np
from scipy.stats import norm
```


**ðŸ§ª Question 1: One-Sample z-Test (Two-Tailed)**

ðŸ“˜ Scenario
A smartphone manufacturer claims that the average battery backup of its phone is 24 hours.

A random sample of 36 phones gives:

â€¢ Sample mean = 23.4 hours

â€¢ Population standard deviation (Ïƒ) = 2 hours

Test the claim at Î± = 0.05.

```python
H0 = 'Battery backup = 24 hours'
H1 = 'Battery backup != 24 hours'

mu = 24
n = 36
x_bar = 23.4
sigma = 2
alpha = 0.05

z_score = (x_bar - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

critical_z = norm.ppf(1 - alpha/ 2)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'Z score: {z_score}')
print(f'P value: {p_value}')
print(f'Critical z: {critical_z}')
```

```python
if (z_score < -critical_z) or (z_score > critical_z):
    print('The z-score falls outside the decision boundary, so we reject the null hypothesis. The battery backup is significantly different from 24 hours. ')
else:
    print('The z-score falls within the decision boundary, so we fail to reject the null hypothesis. The battery backup is NOT significantly different from 24 hours.')
```

```python
if p_value < alpha:
    print('The p-value is less than 0.05, so we reject the null hypothesis. The battery backup is significantly different from 24 hours.')
else:
    print('The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis.The battery backup is NOT significantly different from 24 hours.')
```

**ðŸ§ª Question 2: One-Sample z-Test (Right-Tailed)**

ðŸ“˜ Scenario

An online learning platform claims that the average daily study time of users is at most 4 hours.

A sample of 49 users shows:

â€¢ Sample mean = 4.3 hours

â€¢ Ïƒ = 0.7 hours

Test the claim at Î± = 0.05.


```python
H0 = 'Average study time <= 4 hours'
H1 = 'Average study time > 4 hours'

mu = 4
n = 49
x_bar = 4.3
alpha = 0.05
sigma = 0.7


z_score = (x_bar - mu) / (sigma / np.sqrt(n))
p_value = (1 - norm.cdf(abs(z_score)))

critical_z = norm.ppf(1 - alpha)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'Z score: {z_score}')
print(f'P value: {p_value}')
print(f'Critical z: {critical_z}')
```

```python
if z_score > critical_z:
    print('The z-score falls outside the decision boundary, so we reject the null hypothesis. The average study time is significantly greater than 4 hours. ')
else:
    print('The z-score falls within the decision boundary, so we fail to reject the null hypothesis. The average study time is NOT significantly different from 4 hours.')
```

```python
if p_value < alpha:
     print('The p-value is less than 0.05, so we reject the null hypothesis. The average study time is significantly greater than 4 hours. ')
else:
    print('The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis.The average study time is NOT significantly different from 4 hours.')
```

**ðŸ§ª Question 3: Two-Sample z-Test (Two-Tailed)**

ðŸ“˜ Scenario

Two logistics companies claim to have the same average delivery time.

Company A:
Mean = 42 min, Std = 6 min, n = 64

Company B:
Mean = 39 min, Std = 6 min, n = 81

At Î± = 0.05, test whether there is a significant difference in delivery times.


```python
H0 = 'Company A average delivery time = Company B average delivery time'
H1 = 'Company A average delivery time != Company B average delivery time'

x_bar1 = 42
sigma1 = 6
n1 = 64
x_bar2 = 39
sigma2 = 6
n2 = 81
alpha = 0.05

z_score = (x_bar1 - x_bar2) / np.sqrt((sigma1 ** 2 / n1) + (sigma2 ** 2 / n2))
p_value = 2 * (1 - norm.cdf(abs(z_score)))

critical_z = norm.ppf(1 - alpha /2)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'Z score: {z_score}')
print(f'P value: {p_value}')
print(f'Critical z: {critical_z}')
```

```python
if (z_score < -critical_z) or (z_score > critical_z):
    print('The z-score falls outside the decision boundary, so we reject the null hypothesis. The average delivery time of both companies is significantly different. ')
else:
    print('The z-score falls within the decision boundary, so we fail to reject the null hypothesis. The average delivery time of both companies is NOT significantly different. ')
```

```python
if p_value < alpha:
    print('The p-value is less than 0.05, so we reject the null hypothesis. The average delivery time of both companies is significantly different. ')
else:
    print('The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis.The average delivery time of both companies is NOT significantly different.')
```

**ðŸ§ª Question 4: Two-Sample z-Test (Left-Tailed)**

ðŸ“˜ Scenario

A new software algorithm claims to reduce system response time compared to the old algorithm.

Algorithm Old:
Mean = 8.5 sec, Std = 1.5 sec, n = 100

Algorithm New:
Mean = 7.9 sec, Std = 1.5 sec, n = 90

Test the claim at Î± = 0.05.


```python
H0 = 'Old algorithm response time = New algorithm response time'
H1 = 'Old algorithm response time < New algorithm response time'

x_bar1 = 8.5
sigma1 = 1.5
n1 = 100
x_bar2 = 7.9
sigma2 = 1.5
n2 = 90
alpha = 0.05

z_score = (x_bar1 - x_bar2) / np.sqrt((sigma1 ** 2 / n1) + (sigma2 ** 2 / n2))
p_value = norm.cdf(abs(z_score))

critical_z = norm.ppf(alpha)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'Z score: {z_score}')
print(f'P value: {p_value}')
print(f'Critical z: {critical_z}')
```

```python
if z_score < critical_z:
     print('The z-score falls outside the decision boundary, so we reject the null hypothesis. The new algorithm response time is significantly lower than old algorithm response time.')
else:
    print('The z-score falls within the decision boundary, so we fail to reject the null hypothesis. The new algorithm response time is similar to the old algorithm response time.')

```

```python
if p_value < alpha:
     print('The p-value is less than 0.05, so we reject the null hypothesis. The new algorithm response time is significantly lower than old algorithm response time. ')
else:
    print('The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis. The new algorithm response time is similar to the old algorithm response time.')

```
