## üìå Mini Assignment

Claim: Average delivery time is 30 minutes.
- Define H‚ÇÄ and H‚ÇÅ
- Assume sample data
- Interpret result (conceptual)


```python
import numpy as np
import pandas as pd

np.random.seed(70)
delivery_time = pd.Series(np.random.normal(loc=30, scale=5, size=200))
delivery_time.head()
```

```python
sample_mean = delivery_time.mean()
print(f'The sample mean is: {sample_mean:.2f}')
```

```python
H0 = "Average delivery time = 30"
H1 = "Average delivery time ‚â† 30"

print(f'The H0 is: {H0}')
print(f'The H1 is: {H1}')
```

```python
if sample_mean == 30:
    print(f'Data supports the Null Hypothesis (H0).')
else:
    print(f'Data does NOT supports the Null Hypothesis (H0).')
```
