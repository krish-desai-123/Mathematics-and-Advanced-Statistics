
# 📌 Mini Assignment

Skill vectors:

E1 = [4, 6]

E2 = [5, 8]

Tasks:
1. Compute angle between E1 and E2.
2. Compute projection of E1 onto E2.
3. Visualize vectors and projection.
4. Create 2x2 matrix of skills.
5. Compute determinant and inverse.
6. Explain hyperplane in ML context.
7. Provide real-world interpretation.


```python
import numpy as np
import matplotlib.pyplot as plt

e1 = np.array([4,6])
e2 = np.array([5,8])

dot_prod = np.dot(e1,e2)
mag_e1 = np.linalg.norm(e1)
mag_e2 = np.linalg.norm(e2)

cos_theta = dot_prod / (mag_e1 * mag_e2)
angle = np.degrees(np.arccos(cos_theta))

print(f'Angle: {angle}')
```

```python
plt.quiver(0,0,e1[0],e1[1], angles='xy', scale_units='xy', scale=1, color='red')
plt.quiver(0,0,e2[0],e2[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.xlim(0,10)
plt.ylim(0,10)
plt.title('Angle Between Vectors')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

```python
vector_proj = (dot_prod / (mag_e2**2)) * e2
print(f'Vector Projection: {vector_proj}')
```

```python
plt.quiver(0, 0, vector_proj[0], vector_proj[1], angles="xy", scale_units="xy", scale=1)
plt.quiver(0,0,e1[0],e1[1], angles='xy', scale_units='xy', scale=1, color='red', label='E1')
plt.quiver(0,0,e2[0],e2[1], angles='xy', scale_units='xy', scale=1, color='blue', label='E2')
plt.quiver(0, 0, vector_proj[0], vector_proj[1], angles="xy", scale_units="xy", scale=1, color = 'white', label='Projection')
plt.xlim(0,10)
plt.ylim(0,10)
plt.legend()
plt.title('Vector Projection')
plt.show()
```
