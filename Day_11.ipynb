```python
import numpy as np
from scipy.stats import t, ttest_ind
```

**ðŸ§ª Question 1: One-Sample t-Test (Two-Tailed)**

ðŸ“˜ Scenario

A coaching institute claims that the average mock test score of its students is 70 marks.

A random sample of 12 students gives:
- Sample mean = 72 marks
- Sample standard deviation = 5 marks

Test the claim at Î± = 0.05.

```python
H0 = 'Average mock test score = 70 marks'
H1 = 'Average mock test score != 70 marks'

mu = 70
n = 12
x_bar = 72
alpha = 0.05
s = 5

t_score = (x_bar - mu) / (s / np.sqrt(n))
p_value = 2 * (1 - t.cdf(abs(t_score), df=n-1))

critical_t = t.ppf(1 - alpha/2, df= n-1)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'T score: {t_score}')
print(f'P value: {p_value}')
print(f'Critical t: {critical_t}')
```

```python
if (t_score < -critical_t) or (t_score > critical_t):
    print('The t-score falls outside the decision boundary, so we reject the null hypothesis. The average mock test score is significantly different from 70 marks.')
else:
    print('The t-score falls within the decision boundary, so we fail to reject the null hypothesis. The average mock test score is NOT significantly different from 70 marks.')
```

```python
if p_value < alpha:
     print('The p-value is less than 0.05, so we reject the null hypothesis. The average mock test score is significantly different from 70 marks.')
else:
    print('The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis.The average mock test score is NOT significantly different from 70 marks.')
```

ðŸ§ª **Question 2: One-Sample t-Test (Right-Tailed)**

ðŸ“˜ Scenario

A company claims that the average daily working hours of its employees are more than 8 hours.

A sample of 10 employees shows:
- Sample mean = 8.6 hours
- Sample standard deviation = 0.9 hours

Test the claim at Î± = 0.05.


```python
H0 = 'Average working hour <= 8 hours'
H1 = 'Average working hour > 8 hours'

mu = 8
n = 10
x_bar = 8.6
alpha = 0.05
s = 0.9

t_score = (x_bar - mu) / (s / np.sqrt(n))
p_value = 1 - t.cdf(abs(t_score), df=n-1)

critical_t = t.ppf(1 - alpha, df= n-1)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'T score: {t_score}')
print(f'P value: {p_value}')
print(f'Critical t: {critical_t}')
```

```python
if (t_score < -critical_t) or (t_score > critical_t):
    print('The t-score falls outside the decision boundary, so we reject the null hypothesis. The average working hour is significantly greater than 8 hours.')
else:
    print('The t-score falls within the decision boundary, so we fail to reject the null hypothesis. The average working hour is NOT significantly greater than 8 hours.')
```

```python
if p_value < alpha:
     print('The p-value is less than 0.05, so we reject the null hypothesis. The average working hour is significantly greater than 8 hours.')
else:
    print('The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis.The average working hour is NOT significantly greater than 8 hours.')
```

ðŸ§ª **Question 3: Two-Sample t-Test (Pooled / Equal Variance)**

ðŸ“˜ Scenario

Two batches of students are trained using the same syllabus.

Batch A => Sample Mean=68, Std=4, n=8

Batch B => Sample Mean=64, Std=4, n=8

Assume equal population variances.

Test whether there is a significant difference at Î± = 0.05.

```python
H0 = 'Batch A = Batch B'
H1 = 'Batch A != Batch B'

x_bar1 = 68
x_bar2 = 64
s1 = 4
s2 = 4
n1 = 8
n2 = 8
alpha = 0.05

Batch_A = np.random.normal(x_bar1, s1, n1)
Batch_B = np.random.normal(x_bar2, s2, n2)

t_score, p_value = ttest_ind(Batch_A, Batch_B, equal_var = True)

df = len(Batch_A) + len(Batch_B) - 2

critical_t = t.ppf(1 - alpha, df= df)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'T score: {t_score}')
print(f'P value: {p_value}')
print(f'Critical t: {critical_t}')
```

```python
if (t_score < -critical_t) or (t_score > critical_t):
    print('The t-score falls outside the decision boundary, so we reject the null hypothesis. The mean of the two batches are significantly different.')
else:
    print('The t-score falls within the decision boundary, so we fail to reject the null hypothesis. The mean of the two batches are NOT significantly different.')
```

```python
if p_value < alpha:
    print('The p-value is less than 0.05, so we reject the null hypothesis.The mean of the two batches are significantly different.')
else:
    print('The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis. The mean of the two batches are NOT significantly different.')
```

ðŸ§ª **Question 4: Two-Sample t-Test (Welchâ€™s t-Test)**

ðŸ“˜ Scenario

Two teams are compared based on project completion time (days).


Team A => Mean=15, Std=2, n=6

Team B => Mean=11, Std=4, n=8

Variances are not equal.


```python
H0 = 'Team A = Team B'
H1 = 'Team A != Team B'

x_bar1 = 15
x_bar2 = 11
s1 = 2
s2 = 4
n1 = 6
n2 = 8
alpha = 0.05

Team_A = np.random.normal(x_bar1, s1, n1)
Team_B = np.random.normal(x_bar2, s2, n2)

t_score, p_value = ttest_ind(Team_A, Team_B, equal_var = False)

df = min(len(Team_A)-1 ,len(Team_B)-1)

critical_t = t.ppf(1 - alpha, df= df)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'T score: {t_score}')
print(f'P value: {p_value}')
print(f'Critical t: {critical_t}')
```

```python
if (t_score < -critical_t) or (t_score > critical_t):
    print("The t-score falls outside the decision boundary, so we reject the null hypothesis. The completion times differ significantly.")
else:
    print("The t-score falls within the decision boundary, so we fail to reject the null hypothesis. The completion times do not differ significantly.")
```

```python
if p_value < alpha:
    print("The p-value is less than 0.05, so we reject the null hypothesis. The project completion times differ significantly.")
else:
    print("The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis. There is no significant difference in completion times.")
```

ðŸ§ª **Question 5: Two-Sample t-Test (Left-Tailed)**

ðŸ“˜ Scenario

A new teaching method claims to reduce average learning time compared to the old method.

Method Old => Mean (hrs)=9.5, Std=1.2, n=7

Method New => Mean (hrs)=8.2, Std=1.1, n=7

Test the claim at Î± = 0.05.


```python
H0 = 'Old Method = New Method'
H1 = 'Old Method != New Method'

x_bar1 = 9.5
x_bar2 = 8.2
s1 = 1.2
s2 = 1.1
n1 = 7
n2 = 7
alpha = 0.05

Old = np.random.normal(x_bar1, s1, n1)
New = np.random.normal(x_bar2, s2, n2)

t_score, p_value = ttest_ind(Old, New, equal_var = False)
p_value = p_value / 2

df = min(len(Old)-1 ,len(New)-1)

critical_t = t.ppf(alpha, df= df)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'T score: {t_score}')
print(f'P value: {p_value}')
print(f'Critical t: {critical_t}')
```

```python
if t_score < critical_t:
    print("The t-score falls outside the decision boundary, so we reject the null hypothesis. The new teaching method significantly reduces learning time.")
else:
    print("The t-score falls within the decision boundary, so we fail to reject the null hypothesis. The new teaching method does not significantly reduce learning time.")
```

```python
if p_value < alpha:
    print("The p-value is less than 0.05, so we reject the null hypothesis. The new teaching method significantly reduces learning time.")
else:
    print("The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis. The new method does not significantly reduce learning time.")
```
