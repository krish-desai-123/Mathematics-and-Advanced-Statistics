```python
'''
    Q.2
        - Load a sample dataset (you can use pandas and create it or load from CSV).
        - Use .describe() to get an overview.
        - List the values returned and explain what each one means (e.g., count, mean,and std).
'''

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame(sns.load_dataset('tips'))
df.head()
```

```python
df.describe()
```

```python
# count - Number of non-missing (non-NaN) observations for that column.

# mean -  Gives a central tendency but sensitive to outliers.

# std -   Measures dispersion of values around the mean.

# min -   The minimum (smallest) value in the column.

# 25% -   The 25th percentile (first quartile, Q1).

# 50% -   The 50th percentile (median).

# 75% -   The 75th percentile (third quartile, Q3).

# max -   The maximum (largest) value in the column.
```

```python
'''
    Q.3
        - Calculate the mean of a numeric column using:
            - df['column'].mean()
            - numpy.mean()
        - Find the median using:
            - df['column'].median()
            - numpy.median()
        - Find the mode using:
            - df['column'].mode()
        - Create a bar chart comparing Mean, Median, and Mode.
'''
print(f'Mean (Pandas): {df['total_bill'].mean():.2f}')
print(f'Mean (Numpy): {np.mean(df['total_bill']):.2f}')
print()
print(f'Median (Pandas): {df['total_bill'].median():.2f}')
print(f'Median (Numpy): {np.median(df['total_bill']):.2f}')
print()
print(f'Mode (Pandas): {df['total_bill'].mode()[0]}')
```

```python
sns.barplot(data=df['total_bill'],color='lightblue')
plt.axhline(np.mean(df['total_bill']), color='r', linestyle='--',label='Mean')
plt.axhline(np.median(df['total_bill']), color='b', linestyle='--',label='Median')
plt.axhline(df['total_bill'].mode()[0], color='g', linestyle='--',label='Mode')
plt.legend()
plt.xticks([])
plt.show()
```

```python
'''
    Q.4
        - Calculate the range (max - min) of a numeric column.
        - Calculate variance using:
            - df['column'].var()
            - numpy.var() with ddof=1 for sample variance.
        - Calculate standard deviation using:
            - df['column'].std()
            - numpy.std()
        - Create a boxplot and interpret how dispersion appears visually
'''

range = df['total_bill'].max() - df['total_bill'].min()
print(f'Range: {range}')
print()
print(f'Variance (Pandas): {df['total_bill'].var():.2f}')
print(f'Variance (Numpy): {np.var(df['total_bill'], ddof=1):.2f}')
print()
print(f'Standard Deviation (Pandas): {df['total_bill'].std():.2f})')
print(f'Standard Deviation (Numpy): {np.std(df['total_bill'], ddof=1):.2f})')
```

```python
sns.boxplot(data=df['total_bill'],orient='h', color='red')
plt.title('Dispersion of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Dispersion')
plt.show
```
