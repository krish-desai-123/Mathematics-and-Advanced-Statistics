```python
'''
    Q.3
        - Create a Pandas DataFrame with the following columns:
            - Name (categorical)
            - Age (numerical)
            - Department (categorical)
            - Salary (numerical)
        - Use .dtypes to show data types of each column.
        - Use select_dtypes() to filter numerical and categorical data.
'''

import pandas as pd

df = pd.DataFrame({
    'Name': ['Krish','Kana','Kush','Dev','Parth'],
    'Age': [20,19,21,22,20],
    'Department': ['IT','Sales','Marketing','Admin','HR'],
    'Salary': [89000,67000,90000,120000,72000]
})

print(df)
```

```python
print('\nData types of each column:')
print(df.dtypes)

print('\nCategorical Filter:')
print(df.select_dtypes('str'))

print('\nNumerical Filter:')
print(df.select_dtypes('number'))
```

```python
'''
    Q.4
        - Use Matplotlib or Seaborn to plot:
            - Histogram for age distribution.
            - Bar chart for department count.
            - Pie chart of department percentages.
            - Boxplot for salary.
        - Write 1-2 lines describing the pattern seen in each chart.

'''
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(data=df, x='Age',color='blue')
plt.title('Histogram for age distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

print('\nThis graph represents that the age 20 has higher count, thus indicating most age data falls in that range.')
```

```python
sns.countplot(data=df, x='Department',color='blue')
plt.title('Bar chart for department count')
plt.xlabel('Department')
plt.ylabel('Frequency')
plt.show()

print('\nThis graph represents the count of people who are falling in that department.')
```

```python
plt.pie(x=df['Department'].value_counts(),
        labels = df['Department'].value_counts().index,
        startangle=90,
        colors=['blue','green','red','orange','purple'],
        autopct='%1.1f%%')
plt.title('Pie chart for department count')
plt.xlabel('Department')
plt.show()

print('\nThis graph represents an equal distribution of employees across five departments, with each department accounting for exactly 20% of the total workforce.')
```

```python
import numpy as np

test_scores = np.random.randint(40,101,1000)
test_scores_sample = np.random.choice(test_scores,100)
```

```python
print(f'The population means: {np.mean(test_scores):.2f} & std: {np.std(test_scores):.2f}')
print(f'The sample means: {np.mean(test_scores_sample):.2f} & std: {np.std(test_scores_sample):.2f}')
```
