
# 📌 Mini Assignment

A delivery company tracks vehicle GPS:

V1 = [3, 4]

V2 = [8, 10]

Tasks:
1. Compute distance.
2. Compute L1 and L2 norms.
3. Convert V1 into unit vector.
4. Compute dot product.
5. Interpret cosine similarity.
6. Create 3D vectors and compute cross product.
7. Provide geometric interpretation.


```python
import numpy as np

v1 = np.array([3,4])
v2 = np.array([8,10])

distance = np.linalg.norm(v1 - v2)
print(f'Distance: {distance:.2f}')
```

```python
l1_v1 = np.linalg.norm(v1, ord=1)
l1_v2 = np.linalg.norm(v2, ord=1)

l2_v1 = np.linalg.norm(v1, ord=2)
l2_v2 = np.linalg.norm(v2, ord=2)

print(f'L1 of {v1}: {l1_v1}')
print(f'L2 of {v1}: {l2_v1}')
print()
print(f'L1 of {v2}: {l1_v2}')
print(f'L2 of {v2}: {l2_v2}')
```

```python
mag_v1 = np.linalg.norm(v1)
unit_vector = v1 / mag_v1

mag_unit = np.linalg.norm(unit_vector)

print(f'Unit Vector: {unit_vector}')
print(f'Magnitude of Unit Vector: {mag_unit}')
```

```python
dot_prod = np.dot(v1, v2)
print(f'Dot product of {v1} and {v2}: {dot_prod}')
```

```python
cos_sim = dot_prod / (l2_v1 * l2_v2)
print(f'Cosine similarity of {v1} and {v2}: {cos_sim}')
```

```python
v1 = np.array([3,4,1])
v2 = np.array([8,10,1])

cross = np.cross(v1, v2)
print(f'Cross product of {v1} and {v2}: {cross}')
```
