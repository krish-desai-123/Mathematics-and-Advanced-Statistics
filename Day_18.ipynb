
# 📌 Mini Assignment

A company tracks:
- Experience (years)
- Performance score

Employees:

E1 = [5, 80]

E2 = [7, 85]

E3 = [3, 70]

Tasks:
1. Represent employees as vectors.
2. Compute magnitude of each vector.
3. Compute dot product between E1 and E2.
4. Calculate angle between E1 and E2.
5. Identify most similar employees using cosine similarity.
6. Visualize vectors.
7. Interpret business meaning.


```python
import numpy as np
import math

e1 = np.array([5,80])
e2 = np.array([7,85])
e3 = np.array([3,70])

mag_e1 = np.linalg.norm(e1)
mag_e2 = np.linalg.norm(e2)
mag_e3 = np.linalg.norm(e3)

print(f'E1: {e1}')
print(f'E2: {e2}')
print(f'E3: {e3}')

print(f'Magnitude of e1: {mag_e1}')
print(f'Magnitude of e2: {mag_e2}')
print(f'Magnitude of e3: {mag_e3}')
```

```python
dot_e1_e2 = np.dot(e1,e2)
print(f'Dot product of E1 and E2: {dot_e1_e2}')
```

```python
angle = np.degrees(np.arccos(dot_e1_e2 / (mag_e1 * mag_e2)))
print(f'Angle between E1 and E2: {angle}')
```

```python
def cos_sim(A,B):
    return np.dot(A,B) / (np.linalg.norm(A) * np.linalg.norm(B))

print("Similarity E1-E2:", cos_sim(e1, e2))
print("Similarity E1-E3:", cos_sim(e1, e3))
print("Similarity E2-E3:", cos_sim(e2, e3))
```
