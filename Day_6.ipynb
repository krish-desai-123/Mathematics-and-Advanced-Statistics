## ðŸ“Œ Mini Assignment
Create Student Pass/Fail vs Gender dataset and analyze probabilities.


```python
import pandas as pd
import numpy as np

np.random.seed(90)
student = pd.DataFrame({
    'Gender': np.random.choice(['Male', 'Female'], size=200),
    'Result': np.random.choice(['Pass', 'Fail'], size=200),
})
student.head()
```

```python
prob_male = len(student[student['Gender'] == 'Male']) / len(student)

print(f'Probability of male events: {prob_male}')
```

```python
prob_female = len(student[student['Gender'] == 'Female']) / len(student)

print(f'Probability of female events: {prob_female}')
```

```python
prob_gender = student['Gender'].value_counts(normalize=True)
prob_gender
```

```python
prob_result = student['Result'].value_counts(normalize=True)
prob_result
```

```python
prob_male_pass = len(student[(student['Gender'] == 'Male') & (student['Result'] == 'Pass')]) / len(student)
print(f'Probability of male who passed: {prob_male_pass}')
```

```python
prob_female_pass = len(student[(student['Gender'] == 'Female') & (student['Result'] == 'Pass')]) / len(student)
print(f'Probability of female who passed: {prob_female_pass}')
```

```python
prob_male_fail = len(student[(student['Gender'] == 'Male') & (student['Result'] == 'Fail')]) / len(student)
print(f'Probability of male who failed: {prob_male_fail}')
```

```python
prob_female_fail = len(student[(student['Gender'] == 'Female') & (student['Result'] == 'Fail')]) / len(student)
print(f'Probability of female who failed: {prob_female_fail}')
```

```python
contingency_table = pd.crosstab(student['Gender'], student['Result'])
contingency_table
```

```python
joint_prob_tabel = contingency_table / contingency_table.values.sum()
joint_prob_tabel
```
