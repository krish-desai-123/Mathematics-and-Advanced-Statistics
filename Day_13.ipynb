
## ðŸ“Œ Mini Assignment

A company tests 3 marketing campaigns.

Campaign A: [45, 48, 50, 47]
Campaign B: [52, 55, 53, 54]
Campaign C: [60, 58, 62, 59]


```python
from scipy import stats
import numpy as np
```

```python
H0 = "Mean of Campaign_A, Campaign_B, Campaign_C are equal"
H1 = "Al least one group mean is different "


Campaign_A = np.array([45, 48, 50, 47])
Campaign_B = np.array([52, 55, 53, 54])
Campaign_C = np.array([60, 58, 62, 59])

groups = [Campaign_A, Campaign_B, Campaign_C]

all_data = np.concatenate(groups)
grand_mean = np.mean(all_data)

SSB = sum(len(g) * (np.mean(g) - grand_mean)**2 for g in groups)
SSW = sum(sum((g - np.mean(g))**2) for g in groups)


k = len(groups)
N = len(all_data)

df_between = k - 1
df_within = N - k

MSB = SSB / df_between
MSW = SSW / df_within

F_statistic = MSB / MSW

F_statistic
```

```python
F, p_value = stats.f_oneway(Campaign_A, Campaign_B, Campaign_C)

F, p_value
```

```python
alpha = 0.05
critical_F = stats.f.ppf(1 - alpha, df_between, df_within)

F_statistic, p_value, critical_F
```

```python

if p_value < alpha:
    print("The p-value is less than 0.05, so we reject the null hypothesis. At least one group differs significantly.")
else:
    print("The p-value is greater than or equal to 0.05, so we fail to reject the null hypothesis. No significant difference among group.")
```

```python
if F_statistic > critical_F:
    print("The F-statistic falls outside the decision boundary, so we reject the null hypothesis. At least one group performs differently.")
else:
    print("The F-statistic falls within the decision boundary, so we fail to reject the null hypothesis. Group perform similarly.")
```
