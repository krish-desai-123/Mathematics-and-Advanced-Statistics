```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
scores = pd.read_csv('students_scores.csv')
scores.head()
```

```python
# Maths Score

mean = scores['Math Score'].mean()

median = scores['Math Score'].median()

mode = scores['Math Score'].mode()

print(f'Math Score')
print(f'Mean: {mean}')
print(f'Median: {median}')
print(f'Mode: {mode[0]}')
```

```python
# Science Score

range = scores['Science Score'].max() - scores['Science Score'].min()

var = scores['Science Score'].var()

std = scores['Science Score'].std()

print(f'Science Score')
print(f'Range: {range}')
print(f'Variance: {var:.2f}')
print(f'Standard Deviation: {std:.2f}')
```

```python
# Prob

prob_pass = len(scores[scores['Pass Fail'] == 1]) / len(scores)

contingency_table = pd.crosstab(scores['Pass Fail'], scores['Hours Studied'] > 5)

prob_pass_given_hours_studied_5 = len(scores[(scores['Pass Fail'] == 1) & (scores['Hours Studied'] > 5)]) / len(scores[scores['Hours Studied'] > 5])

print(f'Probability of passing: {prob_pass:.2f}')
print(f'Contingency table: \n{contingency_table}')
print(f'Probability of Passing given hours studied > 5: {prob_pass_given_hours_studied_5:.2f}')
```

```python
# Plot

sns.histplot(scores['Math Score'],kde=True)
plt.title('Histogram of Math Score')
plt.show()
```

```python
skew = scores['Science Score'].skew()
kurto = scores['Science Score'].kurtosis()

print(f'Science Score')
print(f'Skew: {skew:.2f}')
print(f'Kurto: {kurto:.2f}')
```

```python
stats.probplot(scores['English Score'], dist='norm',plot=plt)
plt.title('Q-Q PLot for English Score')
plt.show()
```

```python
# Linear Alg

v_math = scores['Math Score'].head(5).values
v_science = scores['Science Score'].head(5).values
print(f"Math Vector: {v_math}")
print(f"Science Vector: {v_science}")
```

```python
dot_product = np.dot(v_math, v_science)
print(f"Dot Product: {dot_product}")
```

```python
norm_1 = np.linalg.norm(v_math, ord=1)
norm_2 = np.linalg.norm(v_math, ord=2)
print(f"L1 Norm (Math): {norm_1}, L2 Norm (Math): {norm_2:.2f}")
```

```python
cos_theta = dot_product / (np.linalg.norm(v_math) * np.linalg.norm(v_science))
angle_rad = np.arccos(cos_theta)
angle_deg = np.degrees(angle_rad)
print(f"Angle between vectors: {angle_deg:.2f} degrees")
```
