
## ðŸ“Œ Mini Assignment

Claim: Average service time is 5 minutes.
- Assume sample data
- Calculate p-value and CI
- Identify possible errors


```python
import numpy as np
from scipy.stats import norm
np.random.seed(9009)

H0 = 'Average service time = 5 minutes'
H1 = 'Average service time != 5 minutes'

data = np.random.normal(5, 2, 40)
print(data)
```

```python
sample_mean = np.mean(data)
print(sample_mean)
```

```python
mu = 5
x_bar = sample_mean
sigma = 2
n = 40
alpha = 0.05


z_score = (x_bar - mu) / (sigma / np.sqrt(n))
p_value = 2 * (1-norm.cdf(abs(z_score)))

critical_z = norm.cdf(1 - alpha /2)

print(f'H0: {H0}')
print(f'H1: {H1}')
print()
print(f'Z score: {z_score}')
print(f'P value: {p_value}')
print(f'Critical z: {critical_z}')
```

```python
margin = critical_z * (sigma/np.sqrt(n))
lower = x_bar - margin
upper = x_bar + margin
print(f'95% Confidence Interval: {lower:.2f} to {upper:.2f}')
```
