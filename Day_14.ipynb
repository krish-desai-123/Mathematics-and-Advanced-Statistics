**Question 1: Advertising vs Sales (Pearson Correlation)**

A company wants to examine whether Advertising Budget affects Sales.

Advertising (â‚¹ Lakhs): [8, 10, 12, 15, 18, 20, 22, 25]

Sales (â‚¹ Lakhs): [38, 40, 43, 48, 52, 55, 58, 62]

ðŸ“ Tasks
1. Compute covariance.
2. Compute Pearson correlation coefficient.
3. State hypotheses:
- Hâ‚€: r = 0
- Hâ‚: r â‰  0
4. Plot scatter diagram and explain pattern.


```python
import numpy as np
from scipy import stats
import seaborn as sns
ads = np.array([8,10,12,15,18,20,22,25])
sales = np.array([38,40,43,48,52,55,58,62])

covar = np.cov(ads,sales)[0,1]

r, p_value = stats.pearsonr(ads, sales)

print(f'Covariance: {covar}')
print(f'Pearson correlation: {r}')
print(f'P_value: {p_value}')

if r > 0:
    print('Positive Liner Relationship')
elif r < 0:
    print('Negative Liner Relationship')
else:
    print('No Liner Relationship')
```

```python
sns.scatterplot(x=ads, y=sales)
```

**Question 2: Training Hours vs Productivity Rank (Spearman)**

An HR department studies whether Training Hours rank relates to Productivity rank.

Training Rank: [1, 2, 3, 4, 5, 6, 7, 8]

Productivity Rank: [2, 1, 4, 3, 6, 5, 8, 7]


```python
train = np.array([1,2,3,4,5,6,7,8])
prod = np.array([2,1,4,3,6,5,8,7])

r, p_value = stats.spearmanr(train, prod)
print(f'Spearman correlation: {r}')
print(f'P_value: {p_value}')

if r > 0:
    print("Positive Monotonic Relationship")
elif r < 0:
    print("Negative Monotonic Relationship")
else:
    print("No Monotonic Relationship")
```
